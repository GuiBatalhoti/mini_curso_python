## Filtrar colunas e linhas por condicionais

### Uma única condição

Existem vários métodos de filtragem por colunas, como a notação de index padrão, o método loc e o contains.

loc e notação de index são muito parecidos, e assim como o iloc, é mais otimizado. Como o dataset é pequeno não faz tanta diferença. Já o contains tem uma sintaxe um pouco diferente.

```python
dados[dados['Type 1'] == 'Fire'].head()
```

Ao criar a condicional, os dados que serão filtrados serão aqueles que retornaram 'True' apenas

```python
dados['Type 1'] == 'Fire'
```

### Duas condições: AND &

Ainda utilizando a notação de index, todas as vezes que queremos criar duas ou mais condicionais, devemos colocar cada expressão entre parenteses () e conecta-las com algum conectivo lógico

```python
dados[(dados['Type 1'] == 'Grass') & (dados['Type 2'] == 'Poison')]
```

### Duas condições: OR |

repare que o conectivo or é um '|' e o and é um '&' 

```python
## realizando a mesma filtragem mas agora com o operador '|', que representa um 'or'.
dados[(dados['Type 1'] == 'Grass') | (dados['Type 2'] == 'Poison')]
```

### Três condições ou mais

Com tres ou mais condições não é diferente. Segue-se a mesma lógica de isolar as expressões entre parenteses e conectar com um conectivo lógico.

aqui estamos filtrando as linhas onde a cluna 'type 1' é igual a 'grass' e 'type 2' igual a 'poison' e que a coluna  'hp' seja maior que 70


```python
dados2 = dados[(dados['Type 1'] == 'Grass') & (dados['Type 2'] == 'Poison') & (dados['HP'] > 70)]
dados2
```

#### Resetando index

Podemos observar que os indices ficaram estranhos. Então vamos reinicar a contagem dos índices com o método *reset_index*

O parâmetro *drop = True* indica que deve se substituir o índice. Caso o parâmetro seja *drop = False*, seria criado um novo índice e o antigo se tornaria uma outra coluna.

Parâmetro *inplace = True*, em poucas palavas, é uma forma pra não fazer atribuição da variável do dataser novamente (exemplo: *data2 = data2.reset_index()* ).

```python
dados2.reset_index(drop = True, inplace = True)
dados2
```

### Contains

O contains é uma boa alternativa quando queremos filtrar uma substring dentro de uma coluna. A sintaxe pode assustar um pouco no começo mas é bem intuitivo

Aqui estamos filtrando todos os nomes que contém 'Mega'

```python
dados[dados['Name'].str.contains('Mega')].head()
```

#### Operador de negação ~

Como a condicional sempre retorna valores 'True' ou 'False', é possível negar/inverter esses valores com a operação de negação.

Portanto, ao negar a expressão, ele retorna todos os nomes que não contém 'Mega'

```python
dados[~dados['Name'].str.contains('Mega')].head()
```

Ainda é possível utilizar operadores de or ('|') ou and ('&') dentro do contains

Aqui estamos filtrando o atributo 'Type 1' que contém Fire ou Grass. Dessa forma este método tem a mesma função que o filtro anterior

```python
dados[dados['Type 1'].str.contains("Fire|Grass")]

## mas tambem da para fazer do mesmo jeito de antes
#data[(data['Type 1'] == 'Fire') | (data['Type 1'] == 'Grass')]
```

### Loc

#### Alterando um atributo filtrado

a filtragem pelo loc também funciona da mesma forma que as anteriores. Entretanto, uma coisa a mais que é possível fazer é alteração de valores das linhas filtradas de uma forma bem fácil simplesmente especificando a(s) coluna(s) filtradas após a vírgula.

no caso, aqui estamos filtrando as linhas onde a coluna 'type 1' é 'fire' e mudando esses valores para 'flamer'


```python
dados.loc[dados['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
dados.loc[dados['Type 1'] == 'Flamer']
```

#### Alterando mais de um atributo filtrado

utilizando a mesma "formula", é possível modificar mais de uma coluna de uma vez utilizando listas

```python
## para aqueles que tem um total maior que 500, mudamos a coluna 'generation' e 'legendary' para valores de teste
dados.loc[dados['Attack'] > 100, ['Generation', 'Legendary']] = ['TEST 1', 'TEST 2']
dados[dados['Attack'] > 100]
```

## Manipulando Colunas

axis são os eixos que queremos, eixo vertical ou horizontal, como em um plano cartesiano, axis = 1 é vertical e axis = 0 é horizontal

## Valores vazios

### Preencher valores vazios

Preencher valores vazios é extremamente útil. Muitas vezes melhor do que simplesmente apagar a linha. Para isso, usa-se o método fillna().

Mas tem um pequeno problema. Vamos preencher com o que?

Para atributos numéricos, por exemplo, uma estratégia geralmente utilizada é preencher pela média dos valores. Entretanto, isso pode ser muito prejudicial por causa de valores "outliers", ou seja, aqueles valores que são extremamente diferentes em comparação com o restante da distribuição estatística.

Para visualizar melhor, imagina a situação socioeconomica da população mundial. Agora imagina o quanto desses quase 8 bilhões depessoas possuem mais de 1 bilhão de dolares. São muito poucos. Se um bilionário aparecesse aqui mesmo o PIB per capita provavelmente iria subir muito e esse é exatamente o problema da média.

Por isso, uma boa estratégia é preencher com a mediana.

Como aqui não importa muito, vou preencher apenas com uma string aleatória

```python
dados['Type 2'] = dados['Type 2'].fillna('batata')
dados['Type 2']
```

### Remover linhas que possuem valores vazios

Como dito, uma outra estratégia para lidar com valores vazios é simplesmente remover essas linhas. Nem sempre é o mais indicado, principalmente em situações onde há um volume bem pequeno de dados

```python
## usa-se o método dropna() para a remoção de linhas que possuem algum valor vazio
print(len(dados3))
data3 = dados3.dropna()
print(len(dados3))
```

## Amostragem dos dados

Amostragem é você pegar uma fração de um todo.

Para uma boa amostragem, é imprescindível que essa amostra seja o mais diversa e representativa possível. Isso porque em estudos estatísticos, aprendizado de máquina e análise de dados há o problema de viés.

Por exemplo, existem casos de algoritmos de machine learning para reconhecimento facial que são racistas e só reconhecem o rosto de pessoas brancas. Mas como isso pode acontecer sendo que as máquinas não pensam (ainda)? 

Basicamente isso acontece porque os algoritmos foram treinamos com amostras quase que exclusivamente de pessoas brancas. Portanto, hora que ele vê algo diferente disso, ele não consegue reconhecer. Um algoritmo ideal de reconhecimento facil é treinado com a maior diversidade étinica e cultural possível para evitar viéses.

Para fazer amostras em um conjunto de dados, utiliza-se a função sample

```python
## essa fração pode ser definida pelo parâmetro frac
                        ## metade
dados4 = data3.sample(frac = 0.5)
dados4
```

# Matplotlib

Basicamente para plotar todos os graficos é sempre a mesma sequência de passos. Esses passos são

1. Criar a figura

2. Criar o gráfico

3. Configurar o gráfico

4. Mostrar o gráfico

## Importando a biblioteca

```python
import matplotlib.pyplot as plt ## importanto a biblioteca matplotlib e apelindando ela de 'plt'
```

## Grafico de linha

```python


## criando listas x e y de valores para plotar o grafico
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

## criando a figura   
                              ## largura, altura
fig, ax = plt.subplots(figsize = (5,5))

## criando o grafico
ax.plot(x, y)

## configurando o grafico
ax.set(title = 'primeiro plot',
       xlabel = 'x axis',
       ylabel = 'y axis')

## mostrar a figura
plt.show()
```

## Grafico de pontos

```python
## cria a figura
fig, ax = plt.subplots(figsize = (12, 6))

## criando um intervalo
x = range(1, 100, 5) 

f = lambda Xs: [x**2 for x in Xs]

## cria o grafico (scatter é o grafico de pontinhos)
ax.scatter(x, f(x))

## configura o grafico
ax.set(title = 'valores ao quadrado',
        xlabel = 'x',
        ylabel = 'y')

## mostra a figura
plt.show()
```

## Grafico de barras

```python
## plotando barras utilizando um dicionário
produtos = {
    'picanha': 50,
    'contrafile': 30,
    'fraldinha': 40
}

## pega as chaves
x = produtos.keys()
## pega os valores
y = produtos.values()

## cria a figura
fig, ax = plt.subplots()

## cria o grafico (bar é o grafico de barra)
ax.bar(x, y)

## configura o grafico
ax.set(title = 'comparação entre produtos',
       xlabel = 'produtos',
       ylabel = 'preços')

## mostra a figura
plt.show()
```

```python
## plotar a barra na horizontal
fig, ax = plt.subplots()

ax.barh(list(x), list(y))

plt.show()
```

## Histograma

```python
## histograma
import numpy as np

## cria a figura
fig, ax = plt.subplots(figsize = (8,8))

## semente para sempre gerar os mesmos valores "aleatorios"
np.random.seed(42)

## pegar 1000 valores "aleatorios"
x = np.random.randn(1000)

## cria o grafico
ax.hist(x)

## mostra a figura
plt.show()
```

## Multiplos tipos de gráficos em uma única imagem

```python
## cria a figura                      ## duas linhas, duas colunas
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

## cria o(s) graficos
ax[0,0].plot(x, x/2) ## grafico linear
ax[0,1].scatter(np.random.random(10), np.random.random(10)) ## grafico de pontos
ax[1,0].bar(produtos.keys(), produtos.values()) ## grafico de barra
ax[1,1].hist(np.random.randn(1000)) ## histograma

## mostra a figura
plt.show()
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv'

## dados que mostram o preço da gasolina em diferentes países e diferentes anos
dados = pd.read_csv(url)
dados 
```

## multiplos graficos em um único grafico

```python
## cria a figura                   dpi é a densidade de pixels da imagem
fig, ax = plt.subplots(figsize = (6, 6), dpi = 100)

paises_escolhidos = ['Australia', 'USA', 'Canada', 'South Korea']

## é fazer a mesma coisa que plotar vários graficos no mesmo ax
for pais in dados:
  if pais in paises_escolhidos:
    ## cria o(s) grafico(s)
    ax.plot(dados['Year'], dados[pais], marker = '.', label = pais)

## configurando o grafico
ax.set(title = 'preço da gasolina em diferentes países',
       xticks = dados['Year'][::3], ## intervalo do x a cada 3 anos
       xlabel = 'Ano',
       ylabel = 'Valor')

## mostrando a legenda
plt.legend()

## mostrando a figura
plt.show()
```

## Grafico de pizza

```python
import numpy as np

## iniciando variaveis
preco_caro = []
preco_barato = []
precos_2008 = []

## para cada coluna (pais)
for pais in dados:
  ## faz o append dos dados que estão na ultima linha da coluna pais
  precos_2008.append(dados.at[len(dados)-1, pais])

## iterando os preços de 2008
for i in range(1, len(precos_2008)):
  ## se os preços de 2008 for maior que a media
  if precos_2008[i] > np.mean(precos_2008[1:]):
    ## adiciona em gasolina de preço "caro"
    preco_caro.append(precos_2008[i])
  else:
    ## caso contrario, adiciona em gasolina de preço "barato"
    preco_barato.append(precos_2008[i])

## pegando o tamanho da lista de valores que ira compor o grafico, que é a quantidade de paises com gasolinas baratas e caras
precos = [len(preco_caro), len(preco_barato)]
## criando os rotulos
labels = ['% países acima da média', '% países abaixo da média']
## parametro para dar espaçamento entre os valores do grafico
explode = (0.05, 0.05)

## cria a figura
fig, ax = plt.subplots(figsize = (12, 6))

## cria o grafico
ax.pie(precos, labels = labels, autopct = '%.2f', pctdistance=0.8, explode = explode)
## configura o grafico
ax.set(title = f'media dos preços da gasolina em 2008: {round(np.mean(precos_2008), 2)}')
## mostrar legenda
plt.legend()
## mostra a figura
plt.show()
```
